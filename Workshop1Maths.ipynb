Write a function fib (n) that prompts the number of terms “n” from the user and
displays first “n” terms of Fibonacci sequence iteratively.

```python
def fib():
  n = int(input("Enter a term "))
  a = 0
  b = 1
  for i in range (n):
    print(a)
    c = a+b
    a = b
    b = c

fib()

```

Write a function myPow(x, n) that returns 𝑥𝑛, where and n is a non-negative integer.
Do not use the ** operator or the math.pow function - use one while loop

```python
def myPow(x, n):
    result = 1
    i = 0

    while i < n:
        result = result * x
        i = i + 1

    return result


print(myPow(2, 3))
```

Write a function SumDigits(n) that displays the sum of digits of “n”. Hint:
SumDigits(234) will be 2+3+4=9

```python
def SumDigits(n):
    sum = 0

    while n > 0:
        digit = n % 10
        sum = sum + digit
        n = n // 10

    print("Sum of digits =", sum)
SumDigits(234)
```

Write a function squaredSum that returns the sum of the squares of an array/list of
integers such that squaredSum([2,3,4]) gives 29. Write your solution in pure Python
to show the logic in your solution, do not use any library functions.

```python
def squaredSum(arr):
    total = 0

    for num in arr:
        total = total + (num * num)

    return total


print(squaredSum([2,3,4]))
```

Write a function called sumDigits that is given an integer num and returns the sum of
the digits of num. Eg. sumDigits(6) gives 21. Do not use any library functions (if
there is one).


```python
def sumDigits(num):
    total = 0

    while num > 0:
        digit = num % 10
        total = total + digit
        num = num // 10

    return total


print(sumDigits(678))
```
